```r
set.seed(0)
workers <- list()
for (i in 1:400) {
  salary <- sample(7500:30000, 1)
  gender <- sample(c("Male", "Female"), 1)
  worker <- list(id = i, salary = salary, gender = gender)
  workers[[i]] <- worker
}

for (worker in workers) {
  tryCatch({
    if (worker$salary > 10000 && worker$salary < 20000) {
      worker$level <- "A1"
    } else if (worker$salary > 7500 && worker$salary < 30000 && worker$gender == "Female") {
      worker$level <- "A5-F"
    } else {
      worker$level <- "B"
    }
    print(paste("Payment Slip: ID=", worker$id, ", Salary=$", worker$salary, ", Gender=", worker$gender, ", Level=", worker$level, sep=""))
  }, error = function(e) {
    print(paste("Error processing worker ID=", worker$id, ": ", e$message, sep=""))
  })
}
```
